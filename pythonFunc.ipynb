1. What is the difference between a function and a method in Python?

 The Difference Between a Function and a Method in Python Function are

FUNCTION:

* A function is a block of reusable code that performs a specific task.

* It is defined using the def keyword.

* It can be called independently without being tied to any object.

* Example: print(), len(), or your own custom function.

METHOD:

* A method is similar to a function, but it is associated with an object.

* It is called on an object using dot notation.

* Methods automatically receive the object as their first argument (self for instance methods).

* Example: "hello".upper() — here upper() is a method of the string object.

6. Explain the concept of generators in Python and how they are defined.

Generators are a special type of function in Python that produce values one at a time instead of returning all values at once.
They are used to create iterators in a simple and memory-efficient way.

Generators do not store the entire sequence in memory; instead, they generate values on the fly, which makes them useful for large datasets or infinite sequences.

Generators are defined using functions that contain the yield keyword instead of return.

Key characteristics:

Use yield to produce a value.

Execution is paused after each yield and continues from the same point when next called.

Automatically create an iterator.

for example:

    def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1
    gen = count_up_to(3)
    print(next(gen))  # Output: 1
    print(next(gen))  # Output: 2
    print(next(gen))  # Output: 3

Generators Are Useful for the following reasons

* Memory efficient (do not store whole data)

* Faster for large sequences

* Support infinite sequences (e.g., infinite number streams)

* Easy to write compared to creating a full iterator class

9. Explain the purpose and usage of the map() function in Python.

The map() function in Python is used to apply a function to each item of an iterable (such as a list, tuple, or string) and return a new iterator containing the results.

syntax:

    map(function, iterable)

The usage of the map() function:

map() applies a function to every element in an iterable.

It returns an iterator, not a list.

Often used with lambda functions for quick operations.

Helps write cleaner and faster code compared to loops.



It is useful when you want to perform the same operation on every element of a collection without writing a loop.

10. What is the difference between map(), reduce(), and filter() functions in Python?

a.map() — Apply a Function to Every Item

Purpose:

map() applies a function to each element of an iterable and returns a new iterator with the results.

for example:

    numbers = [1, 2, 3]
    result = list(map(lambda x: x * 2, numbers))
    print(result)      # Output: [2, 4, 6]

b.filter() — Keep Items that Meet a Condition
Purpose

filter() selects items from an iterable based on a condition (True/False).

for example:

    numbers = [1, 2, 3, 4, 5]
    result = list(filter(lambda x: x % 2 == 0, numbers))
    print(result)      # Output: [2, 4]

c.reduce() — Reduce All Items to a Single Value
Purpose

reduce() applies a function cumulatively to items in an iterable to produce one final result.

Note: reduce() is available in the functools module.

for example:

    from functools import reduce

    numbers = [1, 2, 3, 4]
    result = reduce(lambda x, y: x + y, numbers)
    print(result)      # Output: 10
